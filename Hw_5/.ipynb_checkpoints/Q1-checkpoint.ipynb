```python
# Plotting libraries
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style='dark')

# Data libraries
import numpy as np
import pandas as pd
import math
```

```python
def stdform(Rarr):
    tconst = []
    tconst1 = []
    tconst2 = []
    parr = []
    for R in Rarr:
        Rc = R+1.001
        timeconst = math.sqrt(0.01/(math.pow(Rc,2)-R*(R+1)))
        
        a = 0.01/(math.pow(Rc,2)-R*(R+1))
        b = 0.2*Rc/(math.pow(Rc,2)-R*(R+1))
        c = 1
        
        root1 = (-b+math.sqrt((b**2)-4*a))/(2*a)
        root2 = (-b-math.sqrt((b**2)-4*a))/(2*a)
        poles = [root1,root2]
        
        z = b/(2*timeconst)
        t1 = timeconst/(z-(math.sqrt((z**2)-1)))
        t2 = timeconst/(z+(math.sqrt((z**2)-1)))
        
        tconst.append(timeconst)
        tconst1.append(t1)
        tconst2.append(t2)
        parr.append(poles)
        
    return [tconst,tconst1,tconst2,parr]
```

```python
R = np.linspace(0,15,1000)
ret = stdform(R)
tconst = ret[0]
tconst1 = ret[1]
tconst2 = ret[2]
poles = ret[3]
poles1 = [i[0] for i in poles]
poles2 = [i[1] for i in poles]
```

```python
# Plot time constant vs R
plt.figure(1,figsize=(15,10))
plt.plot(R,tconst1,label='Time Constant 1')
plt.plot(R,tconst2,label='Time Constant 2')
plt.xlabel('R value')
plt.ylabel('Time Constants')
plt.title('Time Constants vs Recycle Ratio')
plt.legend()
plt.show()
```

```python
# Plot poles
plt.figure(2,figsize=(15,10))
plt.plot(R,poles1,label='First Pole')
plt.plot(R,poles2,label='Second Pole')
plt.xlabel('R value')
plt.ylabel('Pole')
plt.title('Poles vs Recycle Ratio')
plt.legend()
plt.show()
```
